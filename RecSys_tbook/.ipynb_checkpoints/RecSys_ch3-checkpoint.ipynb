## 3.4 Naive Bayes CF
1. que
1. key
    1. Naive Bayes:
        * postier prob: <span style="color:#AF7AC5">$$\begin{aligned} P\left(r_{u j}=v_{s} | \text {Observed ratings in } I_{u}\right) &= \frac{P\left(r_{u j}=v_{s}\right) \cdot P\left(\text {Observed ratings in } I_{u} | r_{u j}=v_{s}\right)}{P\left(\text {Observed ratings in } I_{u}\right)}  \\ & \propto P\left(r_{u j}=v_{s}\right) \cdot P\left(\text {Observed ratings in } I_{u} | r_{u j}=v_{s}\right) \\ & = P\left(r_{u j}=v_{s}\right) \cdot \prod_{k \in I_{u}} P\left(r_{u k} | r_{u j}=v_{s}\right)\end{aligned}$$</span>
        * 2 types of decision:
            1. Maximum Postier Estimation: <span style="color:#AF7AC5">$$\begin{aligned} \hat{r}_{u j} &=\operatorname{argmax}_{v_{s}} P\left(r_{u j}=v_{s} | \text {Observed ratings in } I_{u}\right) \\ &=\operatorname{argmax}_{v_{s}} P\left(r_{u j}=v_{s}\right) \cdot \prod_{k \in I_{u}} P\left(r_{u k} | r_{u j}=v_{s}\right) \end{aligned}$$</span>
            1. weighted average: <span style="color:#AF7AC5">$$ \begin{aligned} \hat{r}_{u j} &=\frac{\sum_{s=1}^{l} v_{s} \cdot P\left(r_{u j}=v_{s} | Observed \text { ratings in } I_{u}\right)}{\sum_{s=1}^{l} P\left(r_{u j}=v_{s} | \text { Observed ratings in } I_{u}\right)} \\ &=\frac{\sum_{s=1}^{l} v_{s} \cdot P\left(r_{u j}=v_{s}\right) \cdot P\left(\text {Observed ratings in } I_{u} | r_{u j}=v_{s}\right)}{\sum_{s=1}^{l} P\left(r_{u j}=v_{s}\right) \cdot P\left(\text {Observed ratings in } I_{u} | r_{u j}=v_{s}\right)} \\ &=\frac{\sum_{s=1}^{l} v_{s} \cdot P\left(r_{u j}=v_{s}\right) \cdot \prod_{k \in I_{u}} P\left(r_{u k} | r_{u j}=v_{s}\right)}{\sum_{s=1}^{l} P\left(r_{u j}=v_{s}\right) \cdot \prod_{k \in I_{u}} P\left(r_{u k} | r_{u j}=v_{s}\right)} \end{aligned} $$</span>            
    1. Reduce overfitting: Laplacian smoothing <span style="color:#AF7AC5">$$ P\left(r_{u j}=v_{s}\right)=\frac{q_{s}+\alpha}{\sum_{t=1}^{l} q_{t}+l \cdot \alpha} $$</span>

## 3.6 Latent Factor Models
1. que:
    1. how the notions of low-rank, redundancy, and correlation are related
1. keys:
    1. intuitiion:
        1. geometric
            1. $\text{low rank} \rightarrow \text{low_dim hyperplane}$
        1. semantic: main point is $R \approx U V^{T}$
        > <img src="img/latent_factor.png" width="50%" height="50%">
    1. basic idea
        1. rotate axis system
        1. representation measured by incomplete data matrix
        1. find a set of latent vectors, minimizing the average squared 
           distance of the data points from the hyperplane defined by 
           these latent vectors.
    1. directly estimate the data matrix in one shot.
    1. general geometric assumption: specified data matrix was available to estimate the relevant latent vector
    1. no correlation means latent factor cannot work well.
    1. residual matrix: noise underlying $R=\left[r_{ij}\right] \in R^{m \times n}$
    1. latent concepts, whether or not semantically interpretable
    1. basic principles:
        1. column --> vector component; row --> factor;
        1. arch: 
        <span style="color:#AF7AC5">$$\begin{aligned} r_{i j} & \approx \overline{u_{i}} \cdot \overline{v_{j}} \\ & \approx \sum_{s=1}^{k} u_{i s} \cdot v_{j s} \\ & =\sum_{s=1}^{k}(\text {Affinity of user i to concept } s) \times(\text {Affinity of item } j \text { to concept } s) \end{aligned}$$</span>
        1. diffs among MF method: 
            1. constrains on $U$ and $V$
            1. nature of objective functions

### 3.6.4 Unconstrained MF:
1. Techinically, UMF $\neq$ SVD
1. optimization form: 
    1. ori: <span style="color:#AF7AC5">$$ \begin{aligned} \text{Minimize } J & =\frac{1}{2}\left\|R-U V^{T}\right\|^{2} \\ &\text{ subject to: No constraints on } U \text{ and } V \end{aligned} $$</span>
    1. update: work on incomplete matrices, but computed only over observed entries: <span style="color:#AF7AC5">$$\begin{aligned} \text { Minimize } J &=\frac{1}{2} \sum_{(i, j) \in S} e_{i j}^{2} \\ &=\frac{1}{2} \sum_{(i, j) \in S}\left(r_{i j}-\sum_{s=1}^{k} u_{i s} \cdot v_{j s}\right)^{2} \\ &\text { subject to: No constraints on } U \text { and } V \end{aligned}$$</span>
1. learning algo: 
    1. gradient descent methods: 
        <span style="color:#AF7AC5">$\begin{array}{l}{U \Leftarrow U+\alpha E V} \\ {V \Leftarrow V+\alpha E^{T} U} \end{array}$</span>
    1. SGD
    1. ILC
    1. ALS/Coordinated descent
1. incorporate other info:
    1. user/item bias
    1. implicit feedback

### 3.6.5 SVD

<span style="color:#AF7AC5">$$ \begin{aligned} \text{Minimize } J & =\frac{1}{2}\left\|R-U V^{T}\right\|^{2} \\ \text{ subject to: } \\ &\text{Columns of } U \text{ are mutually orthogonal} \\ &\text{Columns of } V \text{ are mutually orthogonal} \end{aligned} $$</span>
1. que:
    1. how to optimize
1. basic form: inductive matrix completion

<span style="color:#AF7AC5">$$ \begin{aligned} \text{Minimize } J & =\frac{1}{2}\left\|R-U V^{T}\right\|^{2} \\ \text{ subject to: } \\ & U \geq 0 \\ & V \geq 0 \end{aligned} $$</span>
1. que:
    1. what's the difference between NMF and SVD?
    1. ads and disads
    1. the specific target of NMF
    1. <span style="color:#F4D03F">computation challenge of implicit feedback in 3.6.6.3</span>
1. keys
1. sum:
    1. basic optimization procedure:
        1. zero-filling

        1. add regularization
    1. ads:
        1. interpretability: <span style="color:#AF7AC5">$$\begin{aligned} r_{i j} & \approx \overline{u_{i}} \cdot \overline{v_{j}} \\ & \approx \sum_{s=1}^{k} u_{i s} \cdot v_{j s} \\ & =\sum_{s=1}^{k}(\text {Affinity of user i to concept } s) \times(\text {Affinity of item } j \text { to concept } s) \end{aligned}$$</span>
        1. suitable for non-negative feedbacks, like implicit feedback

    &= \left[\frac{\sum_{l \in Q_{j}(i)} w_{l j}^{i t e m} \cdot\left(r_{i l}-B_{i l}\right)}{\sqrt{\left|Q_{j}(i)\right|}}+\frac{\sum_{l \in Q_{j}(i)} c_{l j}}{\sqrt{\left|Q_{j}(i)\right|}}\right]+\sum_{s=1}^{k+2}\left(u_{i s}+\sum_{h \in I_{i}} \frac{y_{h s}}{\sqrt{\left|I_{i}\right|}}\right) \cdot v_{j s} \end{aligned}$$</span>

        <span style="color:#AF7AC5">$$\begin{array}{l}{\text {Minimize } J=\frac{1}{2} \sum_{(i, j) \in S}\left(r_{i j}-\hat{r}_{i j}\right)^{2}} +\frac{\lambda}{2} \sum_{s=1}^{k+2}\left(\sum_{i=1}^{m} u_{i s}^{2}+\sum_{j=1}^{n} v_{j s}^{2}+\sum_{j=1}^{n} y_{j s}^{2}\right)+\frac{\lambda_{2}}{2} \sum_{j=1}^{n} \sum_{l \in \cup_{i} Q_{j}(i)}\left[\left(w_{l j}^{i t e m}\right)^{2}+c_{l j}^{2}\right]\\ {\text { subject to:}\\ {(k+2) \text {th column of } U \text { contains only } 1 \mathrm{s}}\\ {(k+1) \text {th column of } V \text { contains only } 1 \mathrm{s}}\\{\text { Last two colums of } Y \text { contain only } 0 \mathrm{s}}}\end{array}$$</span>
